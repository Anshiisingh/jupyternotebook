```python
titanic_data= titanic_data.drop(['Cabin'], axis = 1)
```

```python
titanic_data= titanic_data.drop(['Name'], axis = 1)
```

```python
# replacing the missing values in "Age" column with mean value
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
sex_mapping = {"male": 0, "female": 1}
titanic_data['Sex'] = titanic_data['Sex'].map(sex_mapping)
titanic_data.head()
```

```python
titanic_data.replace({'Embarked':{'S':3,'C':4,'Q':5}}, inplace=True)
titanic_data.head(5)
```

```python
titanic_data.head()
```

```python
titanic_data = titanic_data.drop(['Ticket'],axis = 1)
```

```python
titanic_data.head(5)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.dropna().isnull()
```

```python
X = titanic_data.drop(columns = ['PassengerId','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
X_train_prediction
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
X_test_prediction
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
test_data_accuracy
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier

deci = DecisionTreeClassifier()
deci.fit(X_train, Y_train)
y_pred = deci.predict(X_test)
acc_deci = round(accuracy_score(y_pred, Y_test) * 100, 2)
print(acc_deci)
```

```python
##Random forest
from sklearn.ensemble import RandomForestClassifier

randomforest = RandomForestClassifier()
randomforest.fit(X_train, Y_train)
y_pred = randomforest.predict(X_test)
acc_randomforest = round(accuracy_score(y_pred, Y_test) * 100, 2)
print(acc_randomforest)
```
